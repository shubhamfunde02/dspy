## 1. What is need for Hypothesis Testing ? 

### Hypothesis testing is a fundamental concept in Statistics used to make decisions based on data. Its need arises in the following ways:

### 1. Decision-making under uncertainty
### In real-world situations, we often don’t have complete information. Hypothesis testing helps us decide whether an assumption about a population is likely to be true based on sample data.

### 2. Validating assumptions or claims
### It allows researchers to test claims (hypotheses) such as whether a new method, product, or treatment is effective or not.ce
