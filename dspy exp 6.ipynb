## 1. Describe all three data set in terms of following : how many features, their categories (num,text,categorical, nominal,ordinal etc), number of tuples, missing values. 

### The IRIS dataset contains five features: sepal length, sepal width, petal length, petal width, and species. Four of these features are numerical, while species is a categorical nominal feature. The dataset has 150 tuples, and there are no missing values.

### The Air Quality dataset contains 15 features, including various pollutants such as CO, NO₂, O₃, PM2.5, PM10, and meteorological variables like temperature, humidity, and wind speed. Most of the features are numerical, while a few are categorical or ordinal. The number of tuples varies depending on the dataset version, and some columns contain missing values.

### The Wine Quality dataset has 12 features, all of which are numerical except for the quality feature, which is ordinal. The dataset has approximately 4,898 tuples, and it generally does not contain missing values.
